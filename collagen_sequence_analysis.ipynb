```python
def read_seq(inputfile):
    '''Reads and returns de input sequence with special characters removed'''
    with open(inputfile,'r') as f:
        seq = f.read()
        seq = seq.replace('\n','')
        seq = seq.replace('\r','')
    return seq
```

```python
def peptide_count (sequence_1,sequence_2,peptide_lenght):
    '''this function counts the number of times peptides of lenght -peptide_lenght- present on sequence_1 
    occurre on sequence_2. It returns total possible peptides on sequence_1, number of peptides present in both sequences and a list
    of those peptides'''
    total_peptides = 0
    counts = 0
    peptides = []
    i =0
    j = peptide_lenght
    for each in range (len(sequence_1)-peptide_lenght):
        peptide = sequence_1[i:j]
        total_peptides +=1
        if peptide in sequence_2:
            counts+=1
            peptides.append(peptide)
        i+=1
        j+=1
    percentage_of_presence = round(counts*100/total_peptides,2)
    return (counts,percentage_of_presence,total_peptides,peptides)
```
